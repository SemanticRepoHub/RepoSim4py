```python
# Extract codes, docs, requirements, readmes, and structure
repo = "SoftwareUnderstanding/inspect4py"
!mkdir -p {repo} && git clone {f"https://github.com/{repo}.git"} {repo}
!inspect4py -i {repo} -o output/{repo} -dt -sc -rm -r
```

```python
import json

with open(f"output/{repo}/directory_info.json", "r") as f:
    json_info = json.load(f)

# Observing the generated file.
for key, value in json_info.items():
    print(key)
```

```python
# Observing the information of one folder (for extracting codes and docs)
print("The structure of folder:")
folder = json_info.get("output/SoftwareUnderstanding/inspect4py/inspect4py")
print(type(folder)) # list

for element in folder:
    for key, value in element.items():
        print(key)
```

```python
# Observing the information of requirement.
print("The structure of requirements:")
requirements = json_info.get("requirements")
print(type(requirements)) # dict

for key, value in requirements.items():
    print(key)
```

```python
# Observing the information of structure.
print("The structure of directory_tree:")
directory_tree = json_info.get("directory_tree")
print(type(directory_tree)) # dict

for key, value in directory_tree.items():
    print(key)
```

```python
# Observing the information of readme.
print("The structure of readme_files:")
readme_files = json_info.get("readme_files")
print(type(readme_files)) # dict

for key, value in readme_files.items():
    print(key)
```
